```python
#Now divide the 15 into train, evalue set and test set
```

```python
train_size = int(len(data)*.7)
val_size = int(len(data)*.2)+1
test_size = int(len(data)*.1)+1
```

```python
train_size

```

```python
val_size
```

```python
test_size
```

```python
#Now create traindata, valdata and test data using take and skip method fromt the pipline
```

```python
train_data = data.take(train_size)
val_data = data.skip(train_size).take(val_size)
test_data = data.skip(train_size+val_size).take(test_size)
```

```python
    len(train_data)
```

```python
try:
    print(train_data.as_numpy_iterator().next())
except StopIteration:
    print("End of the data")
```

```python
#Now building the model
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Flatten,Dense,Dropout
```

```python
model = Sequential()
```

```python
model.add(Conv2D(16,(3,3),1, activation='relu',input_shape=(256,256,3)))
model.add(MaxPooling2D())

model.add(Conv2D(32,(3,3),1, activation='relu'))
model.add(MaxPooling2D())

model.add(Conv2D(64,(3,3),1, activation='relu'))
model.add(MaxPooling2D())

model.add(Flatten())

model.add(Dense(256,activation='relu'))#Output layer 256-total output present
model.add(Dense(1, activation='sigmoid'))# 1- only on outpur we are going to get
```

```python
model.compile('adam', loss=tf.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
model.summary()
```

```python
 #Now creating logs i.e history of how our model performed...so create a logs directory
```

```python
logdir = 'logs'
```

```python
#tf.keras.callbacks??#Callbacks are useful to get a view on internal states and statistics of the model during training. 
#You can pass a list of callbacks (as the keyword argument callbacks ) to the following model methods: keras. Model.
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

```python
hist = model.fit(train_data,epochs =20, validation_data=val_data, callbacks=[tensorboard_callback])
```

```python
hist.history
```

```python
#Now evaluating the model
```

```python
from tensorflow.keras.metrics import Precision, Recall,BinaryAccuracy
```

```python
pre = Precision()
re = Recall()
acc = BinaryAccuracy()
```

```python
for b in test_data.as_numpy_iterator():
    x,y = b
    yhat = model.predict(x)
    pre.update_state(y,yhat)
    re.update_state(y,yhat)
    acc.update_state(y,yhat)

    
```

```python
print("precision: ",pre.result(),"Recall: ",re.result(),"Accuracy: ",acc.result())
```
